```python
train = pd.read_csv("IPL/train.csv")
```

```python
test = pd.read_csv("IPL/test.csv")
```

```python
test.head()
```

```python
description = pd.read_excel('IPL/data_dictionary.xlsx')
```

```python
# I am droping GameID as i donot see its any significance in prediction
```

```python
train.drop('Game ID',axis=1,inplace=True)
test.drop('Game ID',axis=1,inplace=True)
```

```python
train.columns = list(description['Variable Name'])
test.columns = list(description['Variable Name'])
```

```python
train.head()
```

```python
test.head()
```

```python
#Day do not affect the match
```

```python
train.drop('Day',axis=1,inplace=True)
test.drop('Day',axis=1,inplace=True)
```

```python
#Some team in test data was new to the train data so i am merging them
```

```python
train = pd.concat([train, test], axis=0, sort=False)
```

```python
from sklearn.utils import shuffle
train = shuffle(train)
```

```python
#Adding season because dew affect in bowling and winter we can find dew at night and morning
```

```python
def season(x):
    if int(x[1])<=2 or int(x[1])>=11:
        return 'Winter'
    else:
        return 'Summer'


train['Date of Match'] = train['Date of Match'].str.split('-').apply(season)
train['Date of Match'] = train['Date of Match'].astype('category')
```

```python
#Extracting home city of the team from the team name
```

```python
train['Home_City_1']=train.apply(lambda x: x['City of match'] in x['Team 1'], axis=1)
train['Home_City_2']=train.apply(lambda x: x['City of match'] in x['Team 2'], axis=1)
```

```python
#Categorimg time into evening,morning and afternoon
```

```python
def time(x):
    if int(x[0]) > 16:
        return 'Evening'
    elif int(x[0])<=16 and int(x[0])>=7:
        return 'Afternoon'
    else:
        return 'Morning'
```

team_2

```python
train['Time of Match'] = train['Time of Match'].str.split(':').apply(time)
train['Time of Match'] = train['Time of Match'].astype('category')
```

```python
train['Team won (team 1=1, team 2=0)'] = train['Team won (team 1=1, team 2=0)'].astype('bool')
```

```python
train['Team 1'] = train['Team 1'].astype('category')
train['Team 2'] = train['Team 2'].astype('category')
```

```python
# Making one hot vector representation of categorical data
```

```python
train = pd.get_dummies(train)
train.iloc[:,5:].head(5)
```

```python
# Seperating features and labels to train our models
```

```python
# Use numpy to convert to arrays
import numpy as np
# Labels are the values we want to predict
y = np.array(train['Team won (team 1=1, team 2=0)'])

# Remove the labels from the features
# axis 1 refers to the columns
train= train.drop('Team won (team 1=1, team 2=0)', axis = 1)

# Saving feature names for later use
train_list = list(train.columns)
# Convert to numpy array
train = np.array(train)
```

```python
# Splited into train(80%) and test(20%)
```

```python
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(train, y, test_size = 0.20, random_state = 42)
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier
from sklearn.feature_selection import RFECV
from sklearn.feature_selection import GenericUnivariateSelect
from sklearn.linear_model import RandomizedLogisticRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import RFE
from sklearn import tree
from sklearn.svm import SVR
from sklearn.svm import LinearSVC
from sklearn.metrics import roc_auc_score, accuracy_score, precision_recall_fscore_support
```

```python
# Feature Selection
```

```python
model=Pipeline([('feature_selection', SelectFromModel(LinearSVC(penalty="l2", dual=False))),
    ('classifier', RandomForestClassifier(n_estimators=400, max_depth=10, n_jobs=-1))])
```

```python
model.fit(train_features, train_labels)
prediction = model.predict(test_features)
print( 'accuracy: {}'.format(accuracy_score(test_labels, prediction)))
          
```

```python
# Performance mesasure
```

```python
precision, recall, fscore, support = precision_recall_fscore_support((test_labels),prediction)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))

prob = model.predict_proba(test_features)
prob = [p[1] for p in prob]
print( 'AUROC: {}'.format(roc_auc_score(test_labels, prob)))
```
